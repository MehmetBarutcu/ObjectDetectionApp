```python
config = {}
coord = [p1_cord,p2_cord,p3_cord,p4_cord]
for i in range(len(df)):
    for j,p in enumerate(coord):
        if df.iloc[i][p].values[0][0] != '':
            f_name = df.iloc[i]['kamera_ismi']
            c = []
            for a in df.iloc[i][p].values:
                print(a)
                c.append(tuple(map(int, a)))
            #c = df.iloc[i][p].values.tolist()
            if f_name not in config:
                config[f_name] = {}
            config[f_name][f'p{j}'] = c


    
```

```python
config
```

```python
import json
with open("config.json", "w") as outfile:
    json.dump(config, outfile)
```

```python
with open('config.json','r') as f:
    d = json.load(f)
d
```

```python
for fname in d.keys():
    for cname,c in d[fname].items():
        a = list(map(tuple, c))
        d[fname][cname] = a
print(d)
```
